```python
import pyodbc

# Configurações de conexão
server = 'iron-database-azure.database.windows.net'  # substitua pelo seu servidor
database = 'seu_banco_de_dados'               # substitua pelo seu banco de dados
username = 'seu_usuario'                       # substitua pelo seu usuário
password = 'sua_senha'                         # substitua pela sua senha
driver = '{ODBC Driver 17 for SQL Server}'     # ajuste se necessário

# Conexão com o banco de dados
try:
    conn = pyodbc.connect(
        f'DRIVER={driver};SERVER={server};DATABASE={database};UID={username};PWD={password}'
    )
    print("Conexão bem sucedida!")
except Exception as e:
    print(f"Erro na conexão: {e}")
finally:
    if 'conn' in locals():
        conn.close()
```
